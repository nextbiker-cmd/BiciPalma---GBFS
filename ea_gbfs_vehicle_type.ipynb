```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gbfs_url = "https://gbfs.nextbike.net/maps/gbfs/v2/nextbike_ea/gbfs.json"
response = requests.get(gbfs_url)

if response.status_code == 200:
    gbfs_data = response.json()
else:
    print("Failed to fetch data from the GBFS API")
    exit()
```

```python
#gbfs_data
```

```python
# Assuming 'station_information' is a list, let's iterate through it to find the relevant URL
vehicle_info_url = None
for feed in gbfs_data['data']['en']['feeds']:
    if feed['name'] == 'vehicle_types':
        vehicle_info_url = feed['url']
        break

if vehicle_info_url is None:
    print("Vehicle information URL not found in GBFS data")
    exit()
```

```python
vehicle_info_url
```

```python
# Send GET request
response = requests.get(vehicle_info_url)

if response.status_code == 200:
    vehicle_data = response.json()
else:
    print("Failed to fetch data from the station_info feed")
    exit()
```

```python
vehicle_data
```

```python
vehicle_type=vehicle_data['data']['vehicle_types']
```

```python
vehicle_type
```

```python
# Assuming the 'station_loc' data is in a list format 
vehicles_df = pd.DataFrame(vehicle_type)
vehicles_df
```

```python
vehicles_df.to_csv('vehicle_type_ea.csv', index=False)
```
